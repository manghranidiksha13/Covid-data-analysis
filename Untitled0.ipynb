```python
import pandas as pd
import numpy as np

# Load dataset (assumed available)
df = pd.read_csv("covid_patient_symptoms.csv")

# View basic info
df.info()

# Check missing values
df.isnull().sum()

# Handle missing values (none found, but safe step)
df.fillna(0, inplace=True)

# Correct data types
binary_cols = ['fever', 'dry_cough', 'sore_throat',
               'fatigue', 'headache', 'shortness_of_breath']

df[binary_cols] = df[binary_cols].astype(int)
df['age'] = df['age'].astype(int)
df['gender'] = df['gender'].astype(str)

# Remove duplicate records
df.drop_duplicates(inplace=True)

# Final dataset shape
print(df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.histplot(df['age'], bins=20)
plt.title("Age Distribution of Patients")
plt.show()
```

```python
df['gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.show()
```

```python
symptom_counts = df[binary_cols].sum()

plt.figure()
symptom_counts.plot(kind='bar')
plt.title("Symptom Occurrence Count")
plt.show()
```

```python
plt.figure()
sns.heatmap(df[binary_cols].corr(), annot=True)
plt.title("Symptom Correlation Matrix")
plt.show()
```
